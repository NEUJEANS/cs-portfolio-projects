# Feature pipeline demo
Notebook version of the same preprocessing steps for portfolio screenshots.


```python
def engineer_features(rows):
    output = []
    for row in rows:
        click_rate = min(row['clicks'], 500) / 500
        dwell_rate = min(row['dwell_seconds'], 300) / 300
        output.append({'click_rate': click_rate, 'dwell_rate': dwell_rate, 'converted': row['converted']})
    return output
```
